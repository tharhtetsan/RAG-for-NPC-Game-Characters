```python
import os
from loguru import logger
from repository import VectorRepository
from transform import clean,embed,load
```

```python
file_path = r"./corpus/the_wonderful_wizard_of_oz.txt"

collection_name = "knowledgebase1"
collection_size = 768
chunk_size = 512 

```

```python
# create cllections
vr_repo = VectorRepository()
await vr_repo.create_collections(collection_name=collection_name, size=collection_size)
```

```python
async for chunk in load(filepath=file_path, chunk_size=chunk_size):
        logger.debug(f"Inserting {chunk[0:20]} ... into database")
        embedding_vector = embed(clean(chunk))
        filename = os.path.basename(file_path)
        await vr_repo.create(
            collection_name=collection_name,
            embedding_vector=embedding_vector,
            original_text=chunk,
            source=filename
        )       
```
