## Faiss
Facebook AI Similarity Search (FAISS) is a library for efficient similarity search and clustering of dense vectors. It contains algorithms that search in sets of vectors of any size, up to ones that possibly do not fit in RAM. It also includes supporting code for evaluation and parameter tuning.

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
import os
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings
from langchain_text_splitters import CharacterTextSplitter
```

```python
docs_path = r"/Users/tharhtet/Documents/github/RAG-for-NPC-Game-Characters/corpus/wizard_of_oz/pg54.txt"
```

```python

loader = TextLoader(docs_path)
documents = loader.load()
```

## Hugging Face embedding 

#### FAISS Generate with OPEN AI

```python

#os.environ["OPENAI_API_KEY"] = "YOUR_OPENAI_KEY"
```

```python
# Embedding generate with OpenAI

text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
docs = text_splitter.split_documents(documents)
```

```python
# for openai embedding
openai_embeddings = OpenAIEmbeddings()
db = FAISS.from_documents(docs, openai_embeddings)
```

```python
# for huggingface embeding
from langchain_community.embeddings import HuggingFaceEmbeddings
modelPath = "sentence-transformers/all-MiniLM-l6-v2"
#modelPath = "sentence-transformers/all-mpnet-base-v2"


model_kwargs = {'device':'cpu'}
encode_kwargs = {'normalize_embeddings': False}
embeddings = HuggingFaceEmbeddings(
    model_name=modelPath,     
    model_kwargs=model_kwargs, 
    encode_kwargs=encode_kwargs 
)
```

```python
db = FAISS.from_documents(docs, embeddings)
```

```python
query = "What is my density in this magic land"
query = "Does Dorothy have any friends?"
query = "Where are you from, Dorothy?"
retriever = db.as_retriever()
docs = retriever.invoke(query)
print(docs[0].page_content)
```
